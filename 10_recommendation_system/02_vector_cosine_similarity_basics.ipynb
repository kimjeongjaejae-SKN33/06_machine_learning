## 01. 벡터란 무엇인가?

벡터는 여러 개의 숫자로 대상을 표현한 값이다.

예를 들어 영화 하나를 다음처럼 표현할 수 있다.

- 액션 장르 점수: 1
- 로맨스 장르 점수: 0
- 코미디 장르 점수: 1

그러면 이 영화는 `[1, 0, 1]` 같은 숫자 묶음으로 표현할 수 있다.
추천 시스템에서는 이런 숫자 묶음을 이용해 영화끼리, 상품끼리, 사용자끼리 얼마나 비슷한지 계산한다.


````python
import numpy as np
import pandas as pd

# 각 행은 영화 1개를 의미하고, 각 컬럼은 장르 포함 여부를 의미함.
# 1은 해당 장르가 있음, 0은 해당 장르가 없음을 뜻함.
movie_vector_df = pd.DataFrame({
    'title': ['Movie A', 'Movie B', 'Movie C'],
    'action': [1, 1, 0],
    'romance': [0, 0, 1],
    'comedy': [1, 0, 1]
})

movie_vector_df
````

## 02. 벡터의 크기, 방향, 내적

코사인 유사도를 이해하려면 먼저 벡터의 **크기**, **방향**, **내적**을 구분해야 한다.

벡터는 여러 숫자를 가진 값이지만, 직관적으로는 **화살표**처럼 생각할 수 있다.

- 화살표가 얼마나 긴가? → 벡터의 크기
- 화살표가 어느 쪽을 향하는가? → 벡터의 방향
- 두 화살표가 얼마나 같은 방향으로 놓여 있는가? → 내적과 코사인 유사도로 확인

### 02-01. 벡터의 크기

벡터의 크기는 벡터가 원점에서 얼마나 멀리 뻗어 있는지를 의미한다.

2차원 벡터 $v = (x, y)$의 크기는 다음처럼 계산한다.

$$
||v|| = \sqrt{x^2 + y^2}
$$

예를 들어 벡터 $v = (3, 4)$라면 다음과 같다.

$$
||v|| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5
$$

3차원 벡터 $v = (x, y, z)$라면 다음처럼 확장된다.

$$
||v|| = \sqrt{x^2 + y^2 + z^2}
$$

여기서 $||v||$는 “v의 노름” 또는 “벡터 v의 길이”라고 읽는다.

### 02-02. 벡터의 방향
벡터의 방향은 벡터의 각 성분을 벡터의 크기로 나눈 정규화된 벡터로 구할 수 있다. 이를 단위 벡터(unit vector)라고도 한다.
벡터의 방향은 벡터의 크기를 1로 맞추면 더 쉽게 볼 수 있다.  
벡터를 자기 자신의 크기로 나누면 크기는 1이 되고 방향만 남는다.

$$
\text{unit vector} = \frac{v}{||v||}
$$

예를 들어 $v = (3, 4)$이고 $||v|| = 5$라면 방향은 다음처럼 표현할 수 있다.

$$
\frac{v}{||v||} = \left(\frac{3}{5}, \frac{4}{5}\right) = (0.6, 0.8)
$$

이 값을 단위 벡터라고 한다.

### 02-03. 벡터의 내적

내적은 두 벡터의 같은 위치 숫자끼리 곱한 뒤 모두 더한 값이다.
벡터의 내적을 하는 이유는 두 벡터가 같은 위치에서 얼마나 비슷하게 겹치는지 하나의 숫자로 요약하기 위해서이다.

내적이 크다
= 같은 위치에서 같이 큰 값이 많다
= 두 벡터의 패턴이 비슷할 가능성이 높다


벡터 $A = (a_1, a_2, a_3)$, 벡터 $B = (b_1, b_2, b_3)$가 있으면 내적은 다음과 같다.

$$
A \cdot B = a_1b_1 + a_2b_2 + a_3b_3
$$

예를 들어 $A = (1, 2, 3)$, $B = (4, 5, 6)$이면 다음처럼 계산한다.

$$
A \cdot B = (1 \times 4) + (2 \times 5) + (3 \times 6) = 4 + 10 + 18 = 32
$$

내적은 두 벡터가 같은 위치에서 함께 큰 값을 가질수록 커진다.  
하지만 내적만으로는 두 벡터의 “방향 유사도”를 보기 어렵다. 벡터 자체가 길면 내적도 같이 커질 수 있기 때문이다.

그래서 코사인 유사도에서는 내적을 두 벡터의 크기로 나눈다.


````python
# 내적(행렬곱) 확인
vec_2d = np.array([3,4])
vec_3d = np.array([3,4,5])

# 벡터의 크기 : sqrt((a**2) + (b**2(=))
manual_mag_2d = np.sqrt((3**2) + (4**2))

print("2차원 벡터:", vec_2d)
print("2차원 벡터 크기 직접 계산: ", manual_mag_2d)

# numpy를 이용한 벡터 크기 계산
# linear algebra functions : 선형 대수 함수
norm_2d = np.linalg.norm(vec_2d)
print("2차원 벡터 크기 numpy: ", norm_2d)

norm_3d = np.linalg.norm(vec_3d)
print("3차원 벡터 크기 numpy: ", norm_3d)

# 내적 계산
vec_a = np.array([1,2,3])
vec_b = np.array([4,5,6])

dot_product = np.dot(vec_a, vec_b)
print("내적 연산결과:", dot_product)
````

## 03. 코사인 유사도 수식 유도

코사인 유사도는 두 벡터가 **얼마나 비슷한 방향을 보고 있는지** 계산하는 방법이다.

두 벡터의 내적에는 다음 관계가 있다.

$$
A \cdot B = ||A|| \times ||B|| \times \cos(\theta)
$$

여기서 $\theta$는 두 벡터 사이의 각도이다.

이 식을 $\cos(\theta)$에 대해 정리하면 다음과 같다.

$$
\cos(\theta) = \frac{A \cdot B}{||A|| \times ||B||}
$$
```
cosine similarity = dot product / (벡터 A의 크기 * 벡터 B의 크기)
```

추천 시스템에서는 이 값을 **코사인 유사도(cosine similarity)** 라고 사용한다.



### 03-01. 왜 내적을 크기로 나누는가?

내적은 두 벡터의 값이 함께 클수록 커진다.  
그래서 벡터의 길이가 길면 방향이 비슷하지 않아도 내적이 크게 나올 수 있다.

코사인 유사도는 내적을 $||A|| \times ||B||$로 나누어 크기의 영향을 줄인다.  
그 결과 “값이 얼마나 큰가?”보다 “패턴과 방향이 얼마나 비슷한가?”를 더 잘 볼 수 있다.

### 03-02. 계산 단계 예시

벡터 $A = (1, 2, 3)$, 벡터 $B = (4, 5, 6)$이 있다고 하자.

**1단계: 내적 계산**

$$
A \cdot B = (1 \times 4) + (2 \times 5) + (3 \times 6) = 32
$$

**2단계: 각 벡터의 크기 계산**

$$
||A|| = \sqrt{1^2 + 2^2 + 3^2} = \sqrt{14}
$$

$$
||B|| = \sqrt{4^2 + 5^2 + 6^2} = \sqrt{77}
$$

**3단계: 내적을 두 벡터 크기의 곱으로 나눔**

$$
\text{Cosine Similarity} = \frac{32}{\sqrt{14} \times \sqrt{77}} \approx 0.9746
$$

**4단계: 결과 해석**

`0.9746`은 1에 매우 가까운 값이다.  
따라서 두 벡터는 방향이 매우 비슷하다고 해석할 수 있다.

### 03-03. 값의 의미

- `1`에 가까움: 방향이 거의 같음. 두 대상이 매우 비슷함.
- `0`에 가까움: 방향이 직각에 가까움. 관련성이 낮음.
- `-1`에 가까움: 방향이 반대임.

추천 시스템에서 사용하는 장르 벡터, 단어 카운트 벡터, 평점 벡터는 보통 0 이상의 값으로 구성된다.  
그래서 실제 추천 예제에서는 코사인 유사도가 대체로 `0 ~ 1` 사이에서 해석된다.

주의할 점은 코사인 유사도가 모델 성능 평가 지표가 아니라는 것이다.  
Accuracy, RMSE처럼 정답과 예측을 비교하는 점수가 아니라, 두 아이템 또는 두 벡터가 얼마나 비슷한지 계산하는 기준이다.


````python
def cosine_similarity_manual(a, b):
    # np.dot(): 두 벡터의 같은 위치 숫자끼리 곱한 뒤 모두 더한다.
    # 수식에서 분자에 해당하는 A · B를 계산한다.
    dot_product = np.dot(a, b)

    # np.linalg.norm(): 벡터의 크기를 계산한다.
    # 수식에서 분모에 해당하는 ||A||, ||B||를 계산한다.
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)

    # 코사인 유사도 = 내적 / (벡터 A 크기 * 벡터 B 크기)
    # 벡터 크기 차이를 줄이고 방향과 패턴이 얼마나 비슷한지 비교한다.
    cosine_value = dot_product / (norm_a * norm_b)

    return dot_product, norm_a, norm_b, cosine_value

# 수식 설명에서 사용한 예제 벡터를 그대로 사용한다.
vec_a = np.array([1, 2, 3])
vec_b = np.array([4, 5, 6])

dot_product, norm_a, norm_b, cosine_value = cosine_similarity_manual(vec_a, vec_b)

print('A 벡터:', vec_a)
print('B 벡터:', vec_b)
print('1단계 - 내적 A·B:', dot_product)
print('2단계 - A의 크기 ||A||:', norm_a)
print('2단계 - B의 크기 ||B||:', norm_b)
print('3단계 - ||A|| * ||B||:', norm_a * norm_b)
print('4단계 - 코사인 유사도:', cosine_value)

# 추천 예제에 가까운 장르 벡터도 함께 비교한다.
# Movie A: action과 comedy 속성이 있는 영화
# Movie B: action 속성만 있는 영화
# Movie C: romance와 comedy 속성이 있는 영화
movie_a = np.array([1, 0, 1])
movie_b = np.array([1, 0, 0])
movie_c = np.array([0, 1, 1])

comparison_results = []

for name, target_vector in [('Movie B', movie_b), ('Movie C', movie_c)]:
    dot_product, norm_movie_a, norm_target, cosine_value = cosine_similarity_manual(movie_a, target_vector)

    comparison_results.append({
        'compare': f'Movie A vs {name}',
        'dot_product': dot_product,
        'norm_movie_a': norm_movie_a,
        'norm_target': norm_target,
        'cosine_similarity': cosine_value
    })

comparison_df = pd.DataFrame(comparison_results)
display(comparison_df)
````

## 05. 추천 노트북과의 연결

벡터와 코사인 유사도는 추천 시스템에서 다음처럼 사용된다.

- 콘텐츠 기반 추천: 영화의 장르, 키워드, 줄거리를 숫자 벡터로 바꾼 뒤 비슷한 영화를 찾는다.
- 아이템 기반 협업 필터링: 사용자들이 남긴 평점 패턴을 벡터처럼 보고 비슷한 영화를 찾는다.

즉, 벡터와 코사인 유사도는 추천 시스템의 최종 목표가 아니라, “비슷한 대상을 찾기 위한 계산 도구”이다.
